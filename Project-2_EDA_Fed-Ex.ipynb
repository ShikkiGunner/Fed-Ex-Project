## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Data Wrangling Code

# 1. Basic Info
print("Initial shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)

# 2. Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 3. Convert date columns to datetime
date_cols = ['scheduled_delivery_date', 'delivered_to_client_date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# 4. Create new features
if 'scheduled_delivery_date' in df.columns and 'delivered_to_client_date' in df.columns:
    df['delivery_delay_days'] = (df['delivered_to_client_date'] - df['scheduled_delivery_date']).dt.days


# 5. Handle missing values
df.dropna(subset=['delivery_delay_days'], inplace=True)

# 6. Drop duplicates
df.drop_duplicates(inplace=True)

# 7. Final summary
print("\nAfter wrangling:")
print(df.info())
print("\nSample rows:\n", df.head())
````

### What all manipulations have you done and insights you found?

Standardize column names, Convert date columns to datetime, Create new features, Handle missing values, Removing duplicates.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

# Select numeric columns
numeric_cols = df.select_dtypes(include='number')

# Add delay column if not already
if 'delivery_delay_days' not in numeric_cols.columns and 'delivery_delay_days' in df.columns:
    numeric_cols['delivery_delay_days'] = df['delivery_delay_days']

# Drop rows with NaNs for correlation
numeric_cols = numeric_cols.dropna()

# Compute correlation matrix
corr = numeric_cols.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.show()
````

````python
# Pair Plot visualization code
pairplot_features = ['unit_price', 'line_item_quantity', 'pack_price', 'line_item_value', 'delivery_delay_days']
df_pair = df[pairplot_features].dropna()

# Optional: remove extreme outliers for clarity
df_pair = df_pair[(df_pair['delivery_delay_days'] >= -20) & (df_pair['delivery_delay_days'] <= 60)]

sns.pairplot(df_pair, diag_kind='kde', corner=True)
plt.suptitle("Pair Plot of Key Delivery Features", y=1.02)
plt.tight_layout()
plt.show()
````
